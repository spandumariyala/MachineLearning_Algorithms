```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
feature_cols = ['General_knowledge ', 'Science ']
X = file[feature_cols]
Y = file.SAT
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X,Y)
```

```python
print(lm.intercept_)
print(lm.coef_)
```

```python
list(zip(feature_cols,lm.coef_))
```

```python
Y_predicted = lm.predict(X)
```

```python
Y_predicted
```

```python
file['Y_predicted']=Y_predicted
```

```python
file
```

```python
print('R-Squared:',metrics.r2_score(Y,Y_predicted))
print('Root Mean Squared Error:',np.sqrt(metrics.mean_squared_error(Y,Y_predicted)))
```
